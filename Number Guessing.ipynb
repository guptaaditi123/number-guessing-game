#Number Guessing Game


```python
import random
jackpot = random.randint(1,100)

guess = int(input("Guess the number : "))
Counter = 1
while guess !=jackpot:
  if guess<jackpot:
    print("Guess higher")
  else:
    print("Guess Lower")


  guess = int(input("Guess the number : "))
  Counter +=1
else:
  print("Correct Guess")
  print("Attemps: ", Counter)
```
